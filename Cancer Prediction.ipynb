```python
#Splitting the data

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train,X_test,y_train,y_test=train_test_split(data.data,data.target,random_state=0)

training_accuracy=[]
test_accuracy=[]


for n_neighbors in range(1,11):
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    
    training_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))
    
plt.plot(range(1,11),training_accuracy,label='Training Accuracy')
plt.plot(range(1,11),test_accuracy,label='Test Accuracy')
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.legend()
```

```python
# value of 7 seems to be good fit for n_neighbors.
```

```python
clf=KNeighborsClassifier(n_neighbors=7)
clf.fit(X_train,y_train)
```

## Logistic Regression

```python
# Using Logistic Regression (Linear model of classification)


from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
print('Training score : {}'.format(logreg.score(X_train,y_train)))
print('Test score : {}'.format(logreg.score(X_test,y_test)))
```

```python
#Since training and test scores are close to each other, the model is likely to underfit
```

```python
#Let's try to use more value of C (bias-variance trade off parameter) for more flexible model.

logreg100=LogisticRegression(C=100)
logreg100.fit(X_train,y_train)
```

```python
print('Training score : {}'.format(logreg100.score(X_train,y_train)))
print('Test score : {}'.format(logreg100.score(X_test,y_test)))
```

```python
#Here using c=100 increased the training set accuracy. It means more complex model will perform better
```

```python
logreg1000=LogisticRegression(C=1000).fit(X_train,y_train)
print('Training score : {}'.format(logreg1000.score(X_train,y_train)))
print('Test score : {}'.format(logreg1000.score(X_test,y_test)))
```

```python
# Let's try to use less value of C and see what will happen.

logreg001=LogisticRegression(C=0.01).fit(X_train,y_train)
print('Training score : {}'.format(logreg001.score(X_train,y_train)))
print('Test score : {}'.format(logreg001.score(X_test,y_test)))
```

```python
# Let's look at the coefficients learned by the models

plt.plot(logreg.coef_.T,'*',label='C=1')
plt.plot(logreg001.coef_.T,'o',label='C=0.01')
plt.plot(logreg100.coef_.T,'^',label='C=100')
plt.plot(logreg1000.coef_.T,'v',label='C=1000')

plt.xticks(range(data.data.shape[1]),data.feature_names,rotation=90)
plt.hlines(0,0,data.data.shape[1])
plt.ylim(-7,7)
plt.xlabel('Coefficient index')
plt.ylabel('Coefficient magnitude')
plt.legend()
```

```python
# Stronger regularization pushes coefficients more and more towards zero, though coefficients never become exactly zero.
```

```python
# Here we can see the penalty='l2'. If we want more interpretable model, we must use 'l1-regularization' as it limits the model 
# to using only few important features.
```

```python
C=[0.01,1,100]
for i in C:
    clf_l1=LogisticRegression(C=i,penalty='l1').fit(X_train,y_train)
    print('Training accuaracy with C={:.3f} is {:.3f}'.format(i,clf_l1.score(X_train,y_train)))
    print('Test accuarcy with C={:.3f} is {:.3f}\n'.format(i,clf_l1.score(X_test,y_test)))
```

## Decision Tree Classifier

```python
# Let's use decision tree classifier

from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(random_state=1)
tree.fit(X_train,y_train)
```

```python
print('Training score : {}'.format(tree.score(X_train,y_train)))
print('Test score : {}'.format(tree.score(X_test,y_test)))
```

```python
# Decision tree is more prone to overfitting. We allowed the tree to grow until all the leaves are pure. i.e., each leaf contains
# data points that share the same target value. Hence our model got the score of 100% which indicates overfitting.
```

```python
# Let's apply pre-pruning i.e., stopping the criterion of the tree early

tree1=DecisionTreeClassifier(max_depth=3,random_state=0)
tree1.fit(X_train,y_train)
```

```python
print('Training score : {}'.format(tree1.score(X_train,y_train)))
print('Test score : {}'.format(tree1.score(X_test,y_test)))
```

```python
# We can see the decrease in overfitting now.
```

```python
#Let's visualize the tree

from sklearn import tree
fig=plt.figure(figsize=(25,20))

tree.plot_tree(tree1,class_names=['malignant','benign'],feature_names=data.feature_names,impurity=False,filled=True)
plt.show()
```

```python
fig.savefig('tree.png')
```

```python
# Let's see what all features are important.

print(tree1.feature_importances_)
```

```python
data.data.shape
```

```python
# Let's visualize it.

def important_features(model):
    n_features=data.data.shape[1]
    plt.barh(range(n_features),model.feature_importances_,align='center',height=0.6)
    plt.yticks(np.arange(n_features),data.feature_names)
    plt.xlabel('Feature importance')
    plt.ylabel('Feature')
    
important_features(tree1)
```

# Neural Networks

```python
from sklearn.neural_network import MLPClassifier # Multilayered perceptron
mlp=MLPClassifier(random_state=40)

mlp.fit(X_train,y_train)
```

```python
print('Training score : ',mlp.score(X_train,y_train))
print('Test score : ',mlp.score(X_test,y_test))
```

```python
# We can see the underfitting. It can be resolved if we scale the data. Neural networks expect all the input features to vary in
# similar way, and will require to have mean of 0 and standard deviation 1
```

```python
mean_train=X_train.mean(axis=0)
std_train=X_train.std(axis=0)
X_train_scaled=(X_train-mean_train)/std_train
```

```python
mean_test=X_test.mean(axis=0)
std_test=X_test.std(axis=0)

X_test_scaled=(X_test-mean_test)/std_test
```

```python
mlp1=MLPClassifier(random_state=0)
mlp1.fit(X_train_scaled,y_train)
```

```python
print('Trainig Accuracy : ',mlp1.score(X_train_scaled,y_train))
print('Test Accuracy : ',mlp1.score(X_test_scaled,y_test))
```

```python
# As per the warning, let's increase the number of iterations and let's also increase alpha to add strong regularization to the weights.

mlp2=MLPClassifier(max_iter=300,alpha=1,random_state=0).fit(X_train_scaled,y_train)
```

```python
print('Trainig Accuracy : ',mlp2.score(X_train_scaled,y_train))
print('Test Accuracy : ',mlp2.score(X_test_scaled,y_test))
```
